```python
import h5py
import numpy as np
```

```python
def load_dataset_lazy(filename="offline_dataset.hdf5"):
    f = h5py.File(filename, "r")   # keep file open
    return f   # dict-like interface
```

```python
data = load_dataset_lazy("offline_dataset.hdf5")

obs = data["observations"]        # this is an h5py.Dataset object (not in RAM)
next_obs = data["next_observations"]
print(obs.shape,next_obs.shape)
#single_row = obs[100]             # loaded on-demand (streaming)
```

```python
# check next obs and and obs alignment 

# Check that next_obs[i] == obs[i+1] for all i
all_aligned = np.array_equal(next_obs[:-1], obs[1:])

if all_aligned:
    print("All aligned")
else:
    print("There are misalignments")
```
